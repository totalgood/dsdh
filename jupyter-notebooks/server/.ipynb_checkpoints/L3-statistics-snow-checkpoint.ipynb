```python
base_url = 'https://raw.githubusercontent.com/tscoleman/SnowCholera/master/'
filenames = 'Snow1855_TableIX.csv	Snow1855_TableVIII.csv	Snow1856_TableI.csv	Snow1856_TableVI.csv ' \
            'Snow1855_TableVII.csv	Snow1855_TableXII.csv	Snow1856_TableV.csv '
dfs = []
for filename in filenames.split():
    url = base_url + filename
    print(f'Downloading {url}...')
    # !wget $url
    dfs.append(pd.read_csv(f'../../data/snow/{filename}', skiprows=range(5)))
    display(dfs[-1].describe(include='all'))
```

```python
df_deaths = pd.read_csv('../../data/snow/cholera_deaths.txt')
display(df_deaths.head())
display(df_deaths.describe(include='all'))
```

```python
df_pumps = pd.read_csv('../../data/snow/cholera_pumps.txt')
display(df_pumps.head())
display(df_pumps.describe(include='all'))
```

```python
df_deaths['category'] = 'death'
df_pumps['category'] = 'pump'
df = pd.concat([df_deaths, df_pumps])
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

sns.scatterplot(df.X, df.Y, df.category)
plt.grid('on')
plt.savefig('../../data/snow/snow-scatter.svg', transparent=True)
plt.title('1855 London Cholera Epidemic Analysis by Snow')

plt.show()
```

```python
df_pumps['category'] = 'P'
df_deaths['category'] = 'D'
df = pd.concat([df_deaths, df_pumps])
df.columns = 'x y label'.split()
sns.scatterplot(df['x'], df['y'], df['label'])
plt.grid('on')
# plt.savefig('../../modules/L03/media/snow-scatter-untitled.svg', transparent=True)
plt.show()
```

```python
describe_html = df[df['label']=='P'].describe().round(1).to_html()
with open('../../modules/L03/media/snow-describe-pumps.html', 'w') as fout:
    fout.write(describe_html)
describe_html = df[df['label']=='D'].describe().round(1).to_html()
with open('../../modules/L03/media/snow-describe-deaths.html', 'w') as fout:
    fout.write(describe_html)
```

```python
for orient, a in zip('horizontal vertical'.split(), 'xy'):
    fig = plt.figure()
    for label, group in df.groupby('label'):
        sns.distplot(group[a], bins=range(8, 18), kde=False, label=label)
    plt.legend()
    plt.grid('on')
    plt.savefig(f'../../modules/L03/media/snow-{a}-histogram.svg')
plt.show()
```
