```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display
pd.options.display.max_columns = 500
pd.options.display.max_rows = 30
```

```python
np.random.seed(0)
die = np.floor(6 * np.random.rand(1000)) + 1
sns.distplot(die, bins=[.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5])
plt.grid()
plt.xlabel('dots')
plt.ylabel('probability')
plt.title('1000 rolls, One 6-sided Die')
plt.show()
print('min(die), max(die), np.mean(die), np.median(die), np.std(die) ** 2, np.std(die)')
print(min(die), max(die), np.mean(die), np.median(die), np.std(die) ** 2, np.std(die))
```

```python
np.random.seed(0)
dice = np.sum(np.floor(6 * np.random.rand(1000, 2)) + 1, axis=1)
sns.distplot(dice, bins=np.arange(12) + 1.5)
plt.grid('on')
plt.xlabel('dots')
plt.ylabel('probability')
plt.title('1000 rolls, Two 6-sided Dice')
print('min(dice), max(dice), np.mean(dice), np.median(dice), np.std(dice) ** 2, np.std(dice)')
print(min(dice), max(dice), np.mean(dice), np.median(dice), np.std(dice) ** 2, np.std(dice))
```

```python
np.random.seed(0)
n, m, name = 1000, 3, 'Three'
dice = np.sum(np.floor(6 * np.random.rand(n, m)) + 1, axis=1)
sns.distplot(dice, bins=np.arange(m * 6) + m - .5)
plt.grid('on')
plt.xlabel('dots')
plt.title(f'1000 rolls, {name} 6-sided Dice')
print('min(dice), max(dice), np.mean(dice), np.median(dice), np.std(dice) ** 2, np.std(dice)')
print(min(dice), max(dice), np.mean(dice), np.median(dice), np.std(dice) ** 2, np.std(dice))
```

```python
np.random.seed(0)
# use k-choose-n terminology/parameterization/variable-names
n, m, name, n_sides = 1000000, 3, 'Three', 6
dice = np.sum(np.floor(6 * np.random.rand(n, m)) + 1, axis=1)
sns.distplot(dice, bins=np.arange(m * 6) + m - .5)
plt.grid('on')
plt.xlabel('dots')
plt.title(f'{m} rolls, {name} {n_sides}-sided Dice')
print('min(dice), max(dice), np.mean(dice), np.median(dice), np.std(dice) ** 2, np.std(dice)')
print(min(dice), max(dice), np.mean(dice), np.median(dice), np.std(dice) ** 2, np.std(dice))
```

```python
x = np.random.randn(1000000)
sns.distplot(x, bins=40)
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Normal (Gaussian) Probability Distribution')
plt.xlabel('x = np.random.randn(1 000 000)')
plt.grid('on')
# plt.show()
```
