```python
np.random.seed(0)
die = np.floor(6 * np.random.rand(1000)) + 1
sns.distplot(die, bins=[.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5])
plt.grid()
plt.xlabel('dots')
plt.ylabel('probability')
plt.title('1000 rolls, One 6-sided Die')
plt.show()
print('min(die), max(die), np.mean(die), np.median(die), np.std(die) ** 2, np.std(die)')
print(min(die), max(die), np.mean(die), np.median(die), np.std(die) ** 2, np.std(die))
```

```python
std_dice, mean_dice = {}, {}
```

```python
np.random.seed(0)
# use k-choose-n terminology/parameterization/variable-names
n, m, n_sides = 10000, 3, 6
dice = np.sum(np.floor(n_sides * np.random.rand(n, m)) + 1, axis=1)
bins = np.arange(m * n_sides) + m - .5
sns.distplot(dice, bins=bins)
plt.grid('on')
plt.xlabel('dots')
plt.title(f'{n} rolls of {m} {n_sides}-sided Dice')
std_dice[m] = np.round(np.std(dice), 1)
mean_dice[m] = np.round(np.mean(dice), 1)
stats = pd.Series(name='stats',
    index=f'min(dice), max(dice), np.mean(dice)~={mean_dice[m]}, np.median(dice), np.std(dice)**2, np.std(dice)~={std_dice[m]}'.split(','),
    data=[min(dice), max(dice), np.mean(dice), np.median(dice), np.std(dice)**2, np.std(dice)])
display(stats)
```

```python
np.random.seed(0)
# use k-choose-n terminology/parameterization/variable-names
n, m, n_sides = 10000, 2, 6
dice = np.sum(np.floor(n_sides * np.random.rand(n, m)) + 1, axis=1)
bins = np.arange(m * n_sides) + m - .5
sns.distplot(dice, bins=bins)
plt.grid('on')
plt.xlabel('dots')
plt.title(f'{n} rolls of {m} {n_sides}-sided Dice')
std_dice[m] = np.round(np.std(dice), 1)
mean_dice[m] = np.round(np.mean(dice), 1)
stats = pd.Series(name='stats',
    index=f'min(dice), max(dice), np.mean(dice)~={mean_dice[m]}, np.median(dice), np.std(dice)**2, np.std(dice)~={std_dice[m]}'.split(','),
    data=[min(dice), max(dice), np.mean(dice), np.median(dice), np.std(dice)**2, np.std(dice)])
display(stats)
```
