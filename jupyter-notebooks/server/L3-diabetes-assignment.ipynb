```python
df = pd.read_csv('https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt', sep='\t')
df.head().round(1)
```

```python
columns = [name.lower() for name in df.columns]
columns[-1] = 'severity'
df.columns = columns
pd.options.display.max_rows = 500
display(df.round(1))
pd.options.display.max_rows = 20
```

```python
target_names = ['severity']
feature_names = [name for name in df.columns if name not in target_names]
print(f'feature_names = {feature_names}')
```

Create a training and testset.  
The training set is like the question and answer pairs you get to see during a school lesson.  
The test set is like the exam question and answer pairs that the teacher grades you on at the end of the course.  
Use the training set to show your machine learning model the relationship between your features (age, gender, bmi, blood tests etc) and your target variables (diabetes severity).  
You will use the training set to traiin or fit the model.  
You'll use the test set to see how well you model will work (it's accuracy, standard error, precision, recall, etc) in the real world.  
You'll make predictions for the test set "questions" (`X_test`) and see how closely they match the test set answers (`y_test`).

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
X_train, X_test, y_train, y_test = train_test_split(df[feature_names], df[target_names])
print(f'X_train.shape: {X_train.shape}; y_train.shape: {y_train.shape}')
print(f' X_test.shape: {X_test.shape};  y_test.shape: {y_test.shape}')
```

```python
assert len(X_test) <= len(X_train)   # more data in the training set than the test set
```

```python
y_train = y_train['severity'].values
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
```

```python
lr_bmi = LinearRegression()
lr_bmi = lr_bmi.fit(X_train[['bmi']],
                y_train)
lr_bmi
```

```python
fig = sns.scatterplot(x=X_train['bmi'].values, y=y_train)
print(X_train['bmi'].shape)
print(lr_bmi.coef_)
print(lr_bmi.predict(X_train[['bmi']]).flatten().shape)
fig = plt.plot(X_train['bmi'], lr_bmi.predict(X_train[['bmi']]).flatten(), color='r', linewidth=2)
plt.xlabel('BMI')
plt.ylabel('Diabetes Severity')
plt.title(f'severity = {lr_bmi.coef_.round(2)[0]} * bmi + {lr_bmi.intercept_.round(2)}')
print(f'lr_bmi.intercept_: {lr_bmi.intercept_.round(2)}')
print(f'     lr_bmi.coef_: {lr_bmi.coef_.round(2)}')
```

```python
y_train_pred = lr_bmi.predict(X_train[['bmi']])
# print(f'y_train_pred.shape: {y_train_pred.shape}')
e_train_bmi = y_train_pred - y_train
# print(f'e_train.shape: {e_train_bmi.shape}')
mae_train_bmi = np.sum(np.abs(e_train_bmi)) / len(e_train_bmi)  # .mean()
rmse_train_bmi = np.sqrt((e_train_bmi ** 2).mean())
print(f' mae_train_bmi: {mae_train_bmi.round(1)}')
print(f'rmse_train_bmi: {rmse_train_bmi.round(1)}')
```

```python
y_test_pred = lr_bmi.predict(X_test[['bmi']])
# print(f'y_test_pred.shape: {y_test_pred.shape}')
e_test_bmi = y_test_pred - y_test.values[:,0]
# print(f'e_test.shape: {e_test.shape}')
mae_test_bmi = np.sum(np.abs(e_test_bmi)) / len(e_test_bmi)  # .mean()
rmse_test_bmi = np.sqrt((e_test_bmi ** 2).mean())
print(f' mae_test_bmi: {mae_test_bmi.round(1)}')
print(f'rmse_test_bmi: {rmse_test_bmi.round(1)}')
```

```python
round((rmse_test_bmi - rmse_train_bmi) / rmse_train_bmi, 3)
```

```python
lr_multi = LinearRegression()
lr_multi = lr_multi.fit(X_train, y_train.reshape(-1,1))
print(f'lr_multi.intercept_: {lr_multi.intercept_.round(2)}')
print('lr_multi_coef:')
lr_multi_coef = pd.Series(lr_multi.coef_[0], index=feature_names)
print(lr_multi_coef.round(2))
```

```python
fig = sns.scatterplot(x=X_train['bmi'].values, y=y_train)
print(X_train['bmi'].shape)
print(lr_multi.coef_)
print(lr_multi.predict(X_train).flatten().shape)
fig = sns.scatterplot(X_train['bmi'], lr_multi.predict(X_train).flatten(), color='r', linewidth=2)
plt.xlabel('BMI')
plt.ylabel('Diabetes Severity')
# print(lr_multi_coef['age'])
# print(lr_multi_coef['bmi'])
plt.title(f'severity = ... + {lr_multi_coef["age"].round(2)}*age + {lr_multi_coef["bmi"].round(2)}*bmi + {lr_multi.intercept_.round(2)}')
print(f'lr_multi.intercept_: {lr_multi.intercept_.round(2)}')
print('     lr_multi_coef_:')
print(lr_multi_coef.round(2))
```

```python
y_train_pred = lr_multi.predict(X_train)[:,0]
print(f'y_train_pred.shape: {y_train_pred.shape}')
e_train = (y_train_pred - y_train)
print(f'e_train.shape: {e_train.shape}')
mae_train = np.sum(np.abs(e_train)) / len(e_train)  # .mean()
rmse_train = np.sqrt((e_train ** 2).mean())
print(f' mae_train: {mae_train.round(1)}')
print(f'rmse_train: {rmse_train.round(1)}')
```

```python
y_test_pred = lr_multi.predict(X_test)[:,0]
e_test = y_test_pred - y_test.values[:,0]
mae_test = np.sum(np.abs(e_test)) / len(e_test)  # .mean()
rmse_test = np.sqrt((e_test ** 2).mean())
print(f' mae_test: {mae_test.round(1)}')
print(f'rmse_test: {rmse_test.round(1)}')
```

```python
round((rmse_test - rmse_train) / rmse_train, 4)
```

```python
lr_en = ElasticNet()
lr_en = lr_en.fit(X_train, y_train.reshape(-1,1))
print(f'lr_en.intercept_: {lr_en.intercept_.round(2)}')
print('lr_en_coef:')
lr_en_coef = pd.Series(lr_en.coef_, index=feature_names)
print(lr_en_coef.round(2))
```

```python
y_test_pred = lr_en.predict(X_test)
e_test = y_test_pred - y_test.values[:,0]
mae_test = np.sum(np.abs(e_test)) / len(e_test)  # .mean()
rmse_test = np.sqrt((e_test ** 2).mean())
print(f' mae_test: {mae_test.round(1)}')
print(f'rmse_test: {rmse_test.round(1)}')
```

```python
lr_lasso = Lasso()
lr_lasso = lr_lasso.fit(X_train, y_train.reshape(-1,1))
print(f'lr_lasso.intercept_: {lr_lasso.intercept_.round(2)}')
print('lr_lasso_coef:')
lr_lasso_coef = pd.Series(lr_lasso.coef_, index=feature_names)
print(lr_lasso_coef.round(2))
y_test_pred = lr_lasso.predict(X_test)
e_test = y_test_pred - y_test.values[:,0]
mae_test = np.sum(np.abs(e_test)) / len(e_test)  # .mean()
rmse_test = np.sqrt((e_test ** 2).mean())
print(f' mae_test: {mae_test.round(1)}')
print(f'rmse_test: {rmse_test.round(1)}')
```
